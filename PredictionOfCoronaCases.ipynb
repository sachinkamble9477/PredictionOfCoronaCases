```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
 
# Step One :- Importing and reading file using pandas
df = pd.read_csv('coronadata.csv')
df.head()
 
# Step Two :- Using multi linear regression as the values are continuous where a are independent
#values and b : dependent values
b = df['time']
a = df[['cases','increasingrate','decreasingrate']]
 
# Step Three :- Function named CasesReg for using linear regression model and training the model
# with given values
def CasesReg(a,b):
  from sklearn.linear_model import LinearRegression
  reg = LinearRegression()
  reg.fit(a,b)
  return reg
 
 
# Step Four:- train_test_split is a function in Sklearn model selection for splitting data 
# arrays into two subsets: for training data and for testing data. 
# With this function, you don't need to divide the dataset manually.
a_trn, a_tst, b_trn, b_tst = model_selection.train_test_split(a,b)
 
# Step Five :- Passing training data to my Linear regression model
reg = CasesReg(a_trn,b_trn)
 
# Step Six :- Using predict to get predicted values of b
b_pred = reg.predict(a_tst)
b_pred
 
# Step Seven :- Calculating efficiency. It internally calculates b_pred again and gives the efficiency

reg.score(a_tst,b_tst)
 
# Step Eight :- Plotting graph using matplotlib.
plt.plot(a,reg.predict(a),'*')
plt.legend(labels=['cases','increasingrate','decreasingrate'])
plt.xlabel("Cases Rate")
plt.ylabel("Time")
plt.show()
```
